### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the first dataset [cars1](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv) and [cars2](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv).  

```python
# https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars1.csv
# https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars2.csv
```

   ### Step 3. Assign each to a variable called cars1 and cars2

```python
cars1=pd.read_csv('https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars1.csv')
cars2=pd.read_csv('https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars2.csv')
```

```python
cars1.head()
```

```python
cars2.head()
```

### Step 4. Oops, it seems our first dataset has some unnamed blank columns, fix cars1

```python
# dropping unnamed columns.
# drop function works along column (axis=1)
# the change is permanent (inplace=True)
cars1.drop(['Unnamed: 9',	'Unnamed: 10',	'Unnamed: 11',	'Unnamed: 12',	'Unnamed: 13'], axis=1, inplace=True)
cars1.head()
```

### Step 5. What is the number of observations in each dataset?

```python
cars1.shape[0] # no.of rows in cars1
```

```python
cars2.shape[0] # no.of rows in cars2
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
# joining cars1 and cars2 by column (axis=1)
cars = pd.concat([cars1, cars2], axis=1) 
cars.head()
```

### Step 7. Oops, there is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
# random number series called owners created
# has values ranging from 15000 - 73000
# has as many values as the rows of cars dataframe (cars.shape[0])
owners=np.random.randint(15000, 73000, cars.shape[0])
```

### Step 8. Add the column owners to cars

```python
# added the numpy series to cars as a new column
cars['owners']=owners 
cars.head()
```

###EXTRA. Dropping NULL values

```python
# finally, removing NULL values from the entire dataset
cars.dropna()
```
